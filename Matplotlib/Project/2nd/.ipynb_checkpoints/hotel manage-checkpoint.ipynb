```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hotel = pd.read_csv("hotel_booking_data.csv")
hotel.head()
```

```python
fig,axes = plt.subplots(figsize=(12,6))
scatter = axes.scatter(
            x=hotel["arrival_date_month"],
            y=hotel["required_car_parking_spaces"],
            c=hotel["is_canceled"],
            cmap="winter")

axes.legend(*scatter.legend_elements(),title="Target", loc=(1,0.84));
axes.axhline(y=hotel["required_car_parking_spaces"].mean());

axes.set(
title="Hotel Management system",
xlabel="Arrival date month",
ylabel="Required car parking spaces");
```
