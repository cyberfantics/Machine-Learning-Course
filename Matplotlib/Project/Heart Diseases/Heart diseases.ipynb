```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid'); #updating style
plt.style.available
```

```python
heart = pd.read_csv("heart-disease.csv")
heart.head()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.55,0.5])
axes.set_title("sex")
axes.hist(heart['sex'],label='0=f ,1=m');
axes.legend()

axes = fig.add_axes([0.65,0,0.55,0.5])
axes.set_title("target")
axes.hist(heart['target'],label='0=f ,1=m');
axes.legend()

axes = fig.add_axes([1.3,0,0.55,0.5])
axes.set_title("age")
axes.hist(heart['age'],bins=50);

axes = fig.add_axes([0,-0.75,0.55,0.5])
axes.set_title("oldpeak")
axes.hist(heart['oldpeak'],bins=20);

axes = fig.add_axes([0.65,-0.75,0.55,0.5])
axes.set_title("exang")
axes.hist(heart['exang'],label='0=f ,1=m');
axes.legend()

axes = fig.add_axes([1.3,-0.75,0.55,0.5])
axes.set_title("ca")
axes.hist(heart['ca']);

axes = fig.add_axes([0,-1.5,0.55,0.5])
axes.set_title("cp")
axes.hist(heart['cp']);


axes = fig.add_axes([0.65,-1.5,0.55,0.5])
axes.set_title("fbs")
axes.hist(heart['fbs'],label='0=f ,1=m');
axes.legend()

axes = fig.add_axes([1.3,-1.5,0.55,0.5])
axes.set_title("restecg")
axes.hist(heart['restecg']);


axes = fig.add_axes([0.65,-2.22,0.55,0.5])
axes.set_title("slope")
axes.hist(heart['slope']);

fig.suptitle(x=0.79,y=0.7, t="Heart diseases Histogram\nCreated by Mansoor Buk.");
#fig.savefig("Heart disease histogram.pdf",bbox_inches='tight')
```

```python
over_50 = heart[heart['age']>50]
```

```python
fig,axes = plt.subplots(figsize=(12,7))
scatter = axes.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target']
                      );
axes.set(
    title ="Heart diseases and cholesterol level",
    xlabel = "age",
    ylabel= "cholesterol",
    ylim=(90,500),
    xlim=(50,73),
)

axes.legend(*scatter.legend_elements(),title = "Targets");
axes.axhline(y=heart["chol"].mean(),
             linestyle="--");
```

```python
fig, (axes,axes1) = plt.subplots(figsize=(12,9),nrows=2,ncols=1, sharex=True)
#Base on cholestrol
scatter = axes.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target'],
                     cmap='winter',
                      );
axes.set(
    title ="Heart diseases and cholesterol level",
    #xlabel = "age",
    ylabel= "cholesterol",
    ylim=(100,500),
    xlim=(50,73),
)

axes.legend(*scatter.legend_elements(),title = "Targets");
axes.axhline(y=heart["chol"].mean(),
             linestyle="--");

#Base on heart beat

scatter = axes1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap="winter",
                      );
axes1.set(
    title ="Heart diseases and Max heart rate",
    xlabel = "age",
    ylabel= "heart rate",
    #xlim=(50,75.5)
)

axes1.legend(*scatter.legend_elements(),title="Target")
axes1.axhline(y=heart["thalach"].mean(),
             linestyle="-.");

fig.suptitle("Heart Diseases Analysis By\nMansoor ul Hassan",fontweight='bold',fontsize=30,x=.53)
plt.tight_layout()

#fig.savefig("Heart_diseases_analysis.pdf",bbox_inches='tight')
```
