```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
y=x*2
```

```python
fig = plt.figure(figsize=(5,3),dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label="x,y",marker='o',markersize=4)
axes.plot(x*2,y,label="x^2,y",marker="s",markersize=4)
axes.legend(loc=(0.4,-0.30));
axes.set_title("Graph with two differnt lines",fontsize=18)
axes.set_xlabel("x-axis")
axes.set_ylabel("y-axis")
#fig.savefig("twolines.jpg",bbox_inches='tight')
```
