```python
x = [1, 4, 6, 8]
y = [56, 66, 76, 66]
```

```python
plt.plot(x,y);
```

```python
#1st method
```

```python
fig = plt.figure()
axes = fig.add_subplot()
```

```python
#2nd method
```

```python
fig = plt.figure()
axes = fig.add_subplot()
axes.plot(x,y)
plt.show()
```

```python
#3rd method, this method is most recommended method
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
plt.show()
```

```python
# A simple example for matplotlib
```

```python
fig,axes = plt.subplots(figsize=(5,5))
axes.plot(x,y)
axes.set(title="Simple Plot",
        xlabel='x-axis',
         ylabel='y-axis')

fig.savefig("saved.png")
    
```
