```python
a = np.linspace(0,10,11)
```

```python
b=a**4
```

```python
b
```

```python
a
```

```python
#Add figure and show in graph
fig = plt.figure(figsize=(9,4),dpi=100)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b);
axes1.set_title("Big Graph")
axes1.set_xlabel("x-axis")
axes1.set_ylabel("y-axis")

#Zoom in and add new graph in middle of existing one
axes2 = fig.add_axes([0.3,0.6,0.3,0.3])
axes2.plot(a,b);
axes2.set_title("Zoom in Graph b/w 0,4000")
axes2.set_xlabel("x-axis")
axes2.set_ylabel("y-axis")
axes2.set_xlim(0,8)
axes2.set_ylim(0,4000)


#Zoom in and add new graph in middle of existing one
axes2 = fig.add_axes([0.1,.17,.3,.3])
axes2.plot(a,b);
axes2.set_title("Zoom in Graph b/w 4000,10000")
axes2.set_xlabel("x-axis")
axes2.set_ylabel("y-axis")
axes2.set_xlim(8,10)
axes2.set_ylim(4000,)
#ig.savefig("Graph.png",bbox_inches="tight")
```
