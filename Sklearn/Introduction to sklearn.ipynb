# Make sure that all data is neumerical for this we use car sales to learn

```python
car_sale = pd.read_csv("car-sales-extended.csv")
car_sale.head()
```

```python
X = car_sale.drop('Price',axis=1)
Y=car_sale['Price']
X.head()
```

```python
Y.head()
```

```python
# Turn the catageroies into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

catageroies = ['Make','Colour',"Doors"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot',
                                 one_hot,
                                 catageroies)],
                               remainder = 'passthrough')
transformed_x = transformer.fit_transform(X)
transformed_x = pd.DataFrame(transformed_x)
transformed_x.head()
```
