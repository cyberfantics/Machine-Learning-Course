# Apply() on multiple columns

```python
tip.head(2)
```

```python
def quality(total_bill,tip): 
    if((total_bill/tip)==0.25):
        return "Generous"
    return "Other"
```

```python
tip["Quality"] = tip[["total_bill","tip"]].apply(lambda tip: quality(tip["total_bill"],tip["tip"]),axis = 1) #Both methods are same but it is slower than next ones
```

```python
tip["Quality1"] = np.vectorize(quality)(tip['total_bill'],tip['tip']) #Faster than previous ones
```

```python
tip.head(2)
```

```python
mis = pd.read_csv("movie_scores.csv")
mis
```

```python
mis.isnull() #Checking null values
```

```python
mis.notnull() #checking the values that are not null and having some values
```

```python
mis[mis['pre_movie_score'].notnull()] #Check the data where nothing is missing
```

```python
mis[(mis['pre_movie_score'].isnull() & mis['first_name'].notnull())] ##Check where some values are missing
```

```python
help(mis.dropna)
```

```python
mis.head(1)
```

```python
mis.dropna(thresh=1) #we drop all missing value but that which contain atleast one not null value
```

```python
mis.dropna()
```

```python
mis.dropna(axis=1) #It drop all columns
```

```python
mis.dropna(subset=["sex",'post_movie_score']) #It drop all rows that mis sex and post movie score values
```

```python
mis.fillna(56) #It fill missed values
```

```python
mis['pre_movie_score'].fillna(mis['pre_movie_score'].mean()) #It fill pre missing value with it's mean
```
